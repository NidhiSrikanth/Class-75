```python
import random
import statistics
import pandas as pd
import csv
import plotly.figure_factory as ff
import plotly.graph_objects as go

df= pd.read_csv("data.csv")
data= df["average"].tolist()
#print(df)
#print(data)

fig= ff.create_distplot([data], ["average"], show_hist= False)
fig.show()
```

```python
population_mean= statistics.mean(data)
population_stdev= statistics.stdev(data)
print ("population mean:", population_mean)
print("population stdev:", population_stdev)
```

```python
dataset= []
for i in range(0,100):
  random_index= random.randint(0, len(data))
  value= data[random_index]
  dataset.append(value)

dataset_mean= statistics.mean(dataset)
dataset_stdev= statistics.stdev(dataset)
print("mean of sample data:", dataset_mean)
print("stdev of sample data:", dataset_stdev)
```
